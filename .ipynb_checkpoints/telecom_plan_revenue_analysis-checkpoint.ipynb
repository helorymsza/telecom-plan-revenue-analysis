This classification allows the analysis to focus on active customers while preserving information about churned users for potential future analysis.


```python
# Create customer status column

users['status'] = users['churn_date'].apply(
    lambda x: 'Active' if pd.isna(x) else 'Inactive'
)
```

### 2.3. Calls

```python
# Print general information about 'calls' database

calls.info()
```

```python
# Print a sample of the database

calls.head()
```

### Calls Data Overview

The `calls` DataFrame contains detailed information about phone calls made by Megaline customers during 2018. Each row represents a single call made by a user, identified by a unique call ID.

The dataset includes the user identifier (`user_id`), the date of the call (`call_date`), and the duration of the call in minutes (`duration`). Call durations are recorded as floating-point values, representing the original length of the call before rounding.

According to the business rules of the project, each call is rounded up to the nearest minute for billing purposes. Therefore, the `duration` values will need to be adjusted in later steps to reflect the actual billable minutes.

At this stage, the `call_date` column is stored as a string and should be converted to datetime format to allow for monthly aggregation and time-based analysis. No missing values are present in the dataset, indicating good overall data quality.


### Data Preparation – Calls Dataset

The `call_date` column needs to be converted from string to datetime format to support time-based operations such as grouping calls by month and year.

Additionally, call durations is going to be rounded up to the nearest minute using the ceiling function. This adjustment follows the billing rules of the project, where even very short calls are charged as a full minute. Applying this rule ensures that usage calculations and revenue estimates accurately reflect the business logic of the telecom plans.

No missing values were found in the dataset, and no records were removed during this preparation step.


```python
# Convert call_date to datetime

calls['call_date'] = pd.to_datetime(calls['call_date'])

# Round call duration up to the nearest minute

calls['duration'] = np.ceil(calls['duration'])
```

```python
calls.info()
calls.head()
```

### Agregate data by month

To analyze customer behavior and calculate monthly revenue, call data must be aggregated at the user-month level.
This requires extracting the month from the call date and then summing the total number of minutes used by each user in each month.

Aggregating the data in this way allows us to:

- Compare monthly usage across users

- Apply plan limits correctly

- Calculate overage charges and revenue

```python
# Extract month from call_date

calls['month'] = calls['call_date'].dt.to_period('M') # Converts data into a month period
```

```python
# Aggregate call minutes per user per month

calls_monthly = (
    calls
    .groupby(['user_id', 'month'])
    .agg(total_minutes=('duration', 'sum'))
    .reset_index()
)

calls_monthly.head()
```

### 2.4. Messages

```python
# Print general information about messages database

messages.info()
```

```python
# Print a sample of the database

messages.head()
```

Text messages are billed monthly according to the customer’s plan. Therefore, message data must be aggregated by user and month to calculate monthly usage and later compare it against the plan limits.

Each row in the messages dataset represents a single text message. Since all messages have the same cost unit, we only need to count the number of messages per user per month.

```python
# Convert message data to datetime

messages['message_date'] = pd.to_datetime(messages['message_date'])

# Create month column

messages['month'] = messages['message_date'].dt.to_period('M')

# Aggregate messages per user per month

messages_monthly = (
    messages
    .groupby(['user_id', 'month'])
    .agg(messages_sent=('id', 'count'))
    .reset_index()
)

messages.info()

messages_monthly.head()
```

### 2.5. Internet

```python
# Print general information about internet database

internet.info()
```

```python
# Print a sample of the database

internet.head()
```

Internet usage is billed monthly, and unlike call durations, individual internet sessions are not rounded up.

Instead, the total data usage per month is summed in megabytes and then rounded up to the nearest gigabyte.

Therefore, the internet data must be:

1. Converted to datetime format
2. Aggregated by user and month (sum of MB)
3. Converted from MB to GB
4. Rounded up to the next whole GB

This process ensures that the revenue calculation follows the billing rules defined by Megaline.

```python
# Convert session date to datetime

internet['session_date'] = pd.to_datetime(internet['session_date'])

# Create month column

internet['month'] = internet['session_date'].dt.to_period('M')

internet.info()
```

```python
# Aggregate internet usage per user per month (MB)

internet_monthly = (
    internet
    .groupby(['user_id', 'month'])
    .agg(total_mb=('mb_used', 'sum'))
    .reset_index()
)
```

```python
# Convert MB to GB and round up

internet_monthly['total_gb'] = np.ceil(internet_monthly['total_mb'] / 1024)

internet_monthly.head()
```

### Monthly Internet Usage Aggregation

The internet usage data was aggregated by user and by month in accordance with Megaline’s billing rules. First, session dates were converted to datetime format and a monthly period column was created.

Total internet usage was calculated by summing the number of megabytes used during all sessions within each month. The monthly total was then converted from megabytes to gigabytes and rounded up to the nearest whole gigabyte, as required by the billing policy.

This aggregated dataset will be used later to calculate data overage charges and total monthly revenue per user.


### 3. Monthly Revenue Calculation per User


To calculate the monthly revenue for each customer:

1. Merge the aggregated monthly usage data (calls, messages, internet) with the user’s plan information.
2. Calculate extra usage beyond the plan limits for calls, messages, and internet.
3. Compute extra charges by multiplying overages by the plan rates.
4. Add the monthly subscription fee to get the total revenue per user per month.

This simplified approach ensures each step is easy to understand and follows the project billing rules.


### 3.1. Revenue Calculation

```python
# Merge monthly usage with users to get the plan and city
monthly_usage = calls_monthly.merge(messages_monthly, on=['user_id', 'month'], how='outer')
monthly_usage = monthly_usage.merge(internet_monthly, on=['user_id', 'month'], how='outer')
monthly_usage = monthly_usage.fillna(0)  # fill missing usage with 0
monthly_usage = monthly_usage.merge(users[['user_id', 'plan', 'city']], on='user_id', how='left')
monthly_usage = monthly_usage.merge(plans, left_on='plan', right_on='plan_name', how='left')

# Calculate extra usage
monthly_usage['extra_minutes'] = (monthly_usage['total_minutes'] - monthly_usage['minutes_included']).clip(lower=0)
monthly_usage['extra_messages'] = (monthly_usage['messages_sent'] - monthly_usage['messages_included']).clip(lower=0)
monthly_usage['extra_gb'] = (monthly_usage['total_gb'] - monthly_usage['mb_per_month_included']/1024).clip(lower=0)

# Calculate extra costs
monthly_usage['cost_minutes'] = monthly_usage['extra_minutes'] * monthly_usage['usd_per_minute']
monthly_usage['cost_messages'] = monthly_usage['extra_messages'] * monthly_usage['usd_per_message']
monthly_usage['cost_gb'] = monthly_usage['extra_gb'] * monthly_usage['usd_per_gb']

# Calculate total revenue
monthly_usage['total_revenue'] = (
    monthly_usage['usd_monthly_pay'] +
    monthly_usage['cost_minutes'] +
    monthly_usage['cost_messages'] +
    monthly_usage['cost_gb']
)

# View result
monthly_usage[['user_id', 'month', 'plan', 'total_revenue']].head()
```

### 4. Customer Behavior Analysis by Plan

In this step, we analyze the monthly usage of Megaline customers for each plan. We will examine:

1. Total minutes used per month
2. Number of messages sent per month
3. Total internet usage (in GB) per month

For each plan (Surf and Ultimate), we calculate the mean, variance, and standard deviation.  

Finally, we create histograms to visualize the distribution of usage across customers, which will help identify patterns, outliers, and differences between plans.


```python
# Group by plan
plan_stats = monthly_usage.groupby('plan').agg(
    avg_minutes=('total_minutes', 'mean'),
    var_minutes=('total_minutes', 'var'),
    std_minutes=('total_minutes', 'std'),
    
    avg_messages=('messages_sent', 'mean'),
    var_messages=('messages_sent', 'var'),
    std_messages=('messages_sent', 'std'),
    
    avg_gb=('total_gb', 'mean'),
    var_gb=('total_gb', 'var'),
    std_gb=('total_gb', 'std')
).reset_index()

plan_stats
```

```python
# Set plot style
sns.set_style('whitegrid')

# Create histograms
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Minutes
sns.histplot(data=monthly_usage, x='total_minutes', hue='plan', bins=30, ax=axes[0], kde=False)
axes[0].set_title('Monthly Call Minutes by Plan')
axes[0].set_xlabel('Minutes')
axes[0].set_ylabel('Number of Users')

# Messages
sns.histplot(data=monthly_usage, x='messages_sent', hue='plan', bins=30, ax=axes[1], kde=False)
axes[1].set_title('Monthly Messages Sent by Plan')
axes[1].set_xlabel('Messages')
axes[1].set_ylabel('Number of Users')

# Internet
sns.histplot(data=monthly_usage, x='total_gb', hue='plan', bins=30, ax=axes[2], kde=False)
axes[2].set_title('Monthly Internet Usage (GB) by Plan')
axes[2].set_xlabel('GB')
axes[2].set_ylabel('Number of Users')

plt.tight_layout()
plt.show()
```

### Usage Distributions by Plan

The histograms show how usage differs between the Surf and Ultimate plans:

- **Call minutes**: Ultimate users tend to use more minutes per month, while Surf users mostly stay below the plan limit.
- **Messages**: Ultimate users send significantly more messages on average, reflecting the larger message package.
- **Internet usage**: Ultimate users consume more GB per month, while Surf users have lower usage, often near the plan limit.

The distributions are right-skewed for all three metrics, indicating that a small number of heavy users exceed the plan limits, which will contribute to additional revenue.  
The statistical measures (mean, variance, standard deviation) quantify these differences and support later revenue analysis and hypothesis testing.


### 5. Hypothesis Testing

We will test two hypotheses:

## 1️⃣ Surf vs Ultimate

- **Null hypothesis (H0):** The average monthly revenue of Surf plan users is equal to the average monthly revenue of Ultimate plan users.
- **Alternative hypothesis (H1):** The average monthly revenue of Surf plan users is different from the average monthly revenue of Ultimate plan users.

**Reasoning:** We want to check if the type of plan affects revenue. Since Surf is cheaper and Ultimate includes more minutes, messages, and data, we expect the revenue may differ.

## 2️⃣ NY-NJ vs Other Regions

- **Null hypothesis (H0):** The average monthly revenue of users in NY-NJ-PA MSA is equal to the average monthly revenue of users in other regions.
- **Alternative hypothesis (H1):** The average monthly revenue of users in NY-NJ-PA MSA is different from the average monthly revenue of users in other regions.

**Reasoning:** NY-NJ-PA MSA is a high-density and high-cost region. Testing this hypothesis helps us understand whether regional differe



```python
# -----------------------------
# Hypothesis Test: Surf vs Ultimate
# -----------------------------

from scipy import stats

# Filter the average monthly revenue by plan
revenue_surf = monthly_usage[monthly_usage['plan'] == 'surf']['total_revenue'].dropna()
revenue_ultimate = monthly_usage[monthly_usage['plan'] == 'ultimate']['total_revenue'].dropna()

# Independent two-sample t-test (Welch's t-test)
alpha = 0.05
results = stats.ttest_ind(revenue_surf, revenue_ultimate, equal_var=False)

print("p-value:", results.pvalue)

if results.pvalue < alpha:
    print("Reject the null hypothesis: there is a significant difference in average revenue between Surf and Ultimate plans.")
else:
    print("Fail to reject the null hypothesis: no significant difference in average revenue between Surf and Ultimate plans.")
```

```python
# -----------------------------
# Hypothesis Test: NY-NJ vs Other Regions
# -----------------------------

# Check if 'state' exists in monthly_usage
if 'state' not in monthly_usage.columns:
    # Create 'state' in users if it doesn't exist
    if 'state' not in users.columns:
        users['state'] = users['city'].str.strip()  # keep full city/state string including MSA
    # Merge into monthly_usage
    monthly_usage = pd.merge(monthly_usage, users[['user_id', 'state']], on='user_id', how='left')

# Ensure no leading/trailing spaces
monthly_usage['state'] = monthly_usage['state'].str.strip()

# Filter revenue by region (keep full 'NY-NJ-PA MSA')
revenue_ny_nj = monthly_usage[monthly_usage['state'] == 'NY-NJ-PA MSA']['total_revenue'].dropna()
revenue_other = monthly_usage[monthly_usage['state'] != 'NY-NJ-PA MSA']['total_revenue'].dropna()

print("\nNY-NJ vs Other Regions Revenue Test")
print("NY-NJ sample size:", revenue_ny_nj.shape[0])
print("Other regions sample size:", revenue_other.shape[0])

# Independent two-sample t-test (Welch's t-test)
results_region = stats.ttest_ind(revenue_ny_nj, revenue_other, equal_var=False)

print("p-value:", results_region.pvalue)

if results_region.pvalue < alpha:
    print("Reject the null hypothesis: there is a significant difference in average revenue between NY-NJ and other regions.")
else:
    print("Fail to reject the null hypothesis: no significant difference in average revenue between NY-NJ and other regions.")
```
